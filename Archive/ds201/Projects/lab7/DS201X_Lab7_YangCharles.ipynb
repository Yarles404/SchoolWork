```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
auto = pd.read_csv('https://raw.githubusercontent.com/asukul/DS201/master/datasets/Automobile_price_data_Cleaned.csv')
auto.head()
```

```python
auto.describe()
```

```python
plt.figure(figsize=(10,8))
plt.hist(auto['city-mpg'], bins = 6)
plt.title('DS201X Lab7 YangCharles - Automotive city-mpg Histogram')
plt.grid()

plt.show()
```

```python
auto.boxplot(column = 'city-mpg', by = 'make', figsize = (20, 16), rot = 45, fontsize = 20)

plt.show()
```

```python
auto.plot.scatter(x = 'city-mpg', y = 'engine-size', c = 'curb-weight')

plt.grid()
plt.show()
```
