```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
low_memory = False

file_data = pd.read_csv('Most-Recent-Cohorts-Scorecard-Elements.csv')

file_data.head()

```

```python
file = file_data.groupby(['STABBR']).size()
file.head()
```

```python

ax = file.plot(kind = 'bar', title = '# of Schools per State', figsize = (10, 8), legend = False,
            fontsize = 8, sort_columns = True, color = 'b')
ax.set_xlabel('State', fontsize = 8)
ax.set_ylabel('No of Schools', fontsize = 8)
plt.show()
```
