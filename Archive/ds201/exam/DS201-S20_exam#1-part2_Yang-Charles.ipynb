```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
auto = pd.read_csv('Automobile_price_data_clean-f18.csv')
auto.head()
```

```python
autoMeanPrice = auto.groupby('make').price.mean()
Q1 = autoMeanPrice.plot.bar()
Q1.set_ylabel('Mean Price', color="white")
plt.show() # QUESTION 1: BAR CHART
```

```python
toyota = auto[auto['make'] == 'toyota']
toyota8000 = toyota[toyota['price'] > 8000]

print('Toyota cars priced above $8000 USD: ' + str(len(toyota8000)))
# QUESTION 2
```

```python
hatchback = auto[auto['body-style'] == 'hatchback']
hatchbackMeanCMPG = hatchback['city-mpg'].mean()
print(hatchbackMeanCMPG) # QUESTION 3
```

```python
Q4 = auto.boxplot(column = 'price', by = 'body-style', figsize=(15,10), fontsize=12, rot=45)
Q4.set_ylabel('price')
Q4.set_title('Price Boxplot grouped by Body-style')
plt.show()
```

```python
Q5 = auto.boxplot(column = 'price', by = 'make', figsize=(20,10), fontsize=20, rot=90)
Q5.set_ylabel('price')
Q5.set_title('Price Boxplot grouped by make')
plt.show()
```
