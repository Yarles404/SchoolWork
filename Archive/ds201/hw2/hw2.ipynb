```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
```

```python
LO = pd.read_csv('Lemonade2016-2.csv')
LO
```

```python
LO.isnull()
```

```python
LO.loc[8, 'Date'] = r'7/8/2018'

LO_LeafMean = LO.groupby(['Location']).mean()
#LOLeafMean

LO.loc[20, 'Leaflets'] = int(LO_LeafMean.loc['Park', 'Leaflets'])
LO
```

```python
LO.drop_duplicates()
```

```python
LO['Sales'] = LO['Lemon'] + LO['Orange']
LO['Revenue'] = LO['Sales'] * LO['Price']
LO
```

```python
LO_scatter = LO.plot(kind='scatter', x ='Leaflets', y='Sales')
```

```python
LO_hist = LO.hist('Revenue') #Right Skewed
```

```python
LO_boxplot = LO.boxplot(column = 'Sales', by = 'Location', grid = False)
LO_boxplot.set_ylabel('Sales')
```

```python
LO_Temp75 = LO[LO['Temperature'] >= 75]
LO_Temp75_Beach = LO_Temp75[LO_Temp75['Location'] == 'Beach']
LO_Temp75_Park = LO_Temp75[LO_Temp75['Location'] == 'Park']

fig = plt.figure(1, figsize = (12,8))
ax = fig.add_subplot()
bp = ax.boxplot([LO_Temp75_Park['Sales'], LO_Temp75_Beach['Sales']])
#bp2 = ax.boxplot(LO_Temp75_Beach['Sales'])

ax.set_xticklabels(['Park', 'Beach'], fontsize = 15)
ax.set_ylabel('Sales', fontsize = 15)
ax.set_title('Sales at Park and Beach when Temperature >= 75', fontsize = 20)
```

```python
sb.pairplot(LO, vars = ['Revenue', 'Temperature', 'Leaflets'], hue = "Location")
```
