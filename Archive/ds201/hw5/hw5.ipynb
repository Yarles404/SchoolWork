```python
from bokeh.plotting import figure, show, output_notebook, output_file, save
import pandas as pd
from bokeh.models import ColumnDataSource, CDSView, GroupFilter
from bokeh.layouts import gridplot
```

```python
lemon = pd.read_csv("https://raw.githubusercontent.com/asukul/DS201/master/datasets/Lemonade2016-2.csv")
lemon.drop_duplicates(subset = "Date", inplace = True)
lemon.info()
```

```python
lemon.isna()
```

```python
lemon.loc[8, ['Date']] = "7/8/2016"
lemon.drop(20, inplace = True)
lemon.info()
```

```python
lemon["Sales"] = lemon["Lemon"] + lemon["Orange"]
lemon["Revenue"] = lemon["Sales"] * lemon["Price"]
lemon.info()
```
